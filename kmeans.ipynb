```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X,y=make_blobs(random_state=1)

kmeans=KMeans(n_clusters=3)
kmeans.fit(X)

#kmeans.labels_



import matplotlib.pyplot as plt

y_means=kmeans.predict(X)
plt.scatter(X[:,0],X[:,1], c=y_means)

centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='r', s=100)

plt.xticks(())
plt.yticks(())
```

```python
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram , ward

X,y= make_blobs(random_state=0)

linkage_array=ward(X)
dendrogram(linkage_array)

from matplotlib import pyplot as plt
ax=plt.gca()
bounds=ax.get_xbound()
ax.plot(bounds, [18,18],'--',c='k')
ax.plot(bounds, [10,10],'--',c='k')
ax.text(bounds[1], 18, 'two clusters', va='center')
fontdict={'size':15}
ax.text(bounds[1], 10, 'three clusters', va='center')
fontdict={'size':15}
plt.xlabel('sample index')
plt.ylabel('cluster distance')
```

```python
#iris dataset
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()
X=iris.data
y=iris.target

from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3,random_state=1)
kmeans.fit(X)

y_pred=kmeans.predict(X)

import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y_pred)

centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='r', s=100)
plt.title('KMeans Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.xticks(())
plt.yticks(())
```

```python

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
# Load Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)  # Features
y = pd.DataFrame(iris.target)  # True labels (optional, for comparison later)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=1)
y_means = kmeans.fit_predict(X_scaled)

# Scatter plot: comparing clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_means)  # Using Sepal Length and Sepal Width


centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='r')
plt.title('KMeans Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.xticks(())
plt.yticks(())
```
